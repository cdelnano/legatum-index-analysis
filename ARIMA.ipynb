```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# Load the CSV file into a DataFrame

#df_2019 = pd.read_csv('2019_scores.csv')
df_2020 = pd.read_csv('2020_scores_2_countries.csv')
df_2021 = pd.read_csv('2021_scores_2_countries.csv')
df_2023 = pd.read_csv('2023_scores_2_countries.csv')


# soft data in alphabetical order by country name
#df_2019 = df_2019.sort_values(by='Country')
df_2020 = df_2020.sort_values(by='Country')
df_2021 = df_2021.sort_values(by='Country')
df_2023 = df_2023.sort_values(by='Country')

# Define the columns to analyze
columns_to_analyze = ['Safety & Security', 'Personal Freedom', 'Governance', 'Social Capital',
                      'Investment Environment', 'Enterprise Conditions', 'Market Access & Infrastructure',
                      'Economic Quality', 'Living Conditions', 'Health', 'Education', 'Natural Environment']

# List to store forecasted data for all countries and factors
forecasted_data = []

# Iterate through each country and perform ARIMA modeling for each column
for index, row in df_2021.iterrows():
    country = row['Country']

    years = [2020, 2021, 2023]

    #needs modified to grab values from other df for same country and pillar
    country_values = {}

    # Iterate over each column to analyze
    for column in columns_to_analyze:

        # TO DO: modify to grab score for one country for each year and put into dict
     #   values_2019 = df_2019.loc[df_2019['Country'] == country, column].values
        values_2020 = df_2020.loc[df_2020['Country'] == country, column].values
        values_2021 = df_2021.loc[df_2021['Country'] == country, column].values
        values_2023 = df_2023.loc[df_2023['Country'] == country, column].values
        # Combine values from all years into a single list
      #  combined_values = values_2019.tolist() + values_2020.tolist() + values_2021.tolist() + values_2023.tolist()
        combined_values = values_2020.tolist() + values_2021.tolist() + values_2023.tolist()
        # Create a dictionary entry with column name as key and combined values as value
        country_values[column] = combined_values
    print(country)
    print(country_values)
    # Convert the values to time series data and perform ARIMA modeling for each column
    for column, values in country_values.items():
        ts = pd.Series(values, index=years)
        order = (1, 0, 0)  
        model = ARIMA(ts, order=order)
        results = model.fit()

        # Make predictions using the ARIMA models
        forecast_years = 5  # Number of years to forecast
        forecast = results.forecast(steps=forecast_years)

        # Store the forecasted values for each country and column
        for year in range(forecast_years):
            forecasted_data.append({
                'Country': country,
                'Factor': column,
                'Year': years[-1] + year + 1,  # Adjust year to start from the next year after the last data point
                'Forecasted Value': forecast.iloc[year]
            })
# Convert the forecasted data to a DataFrame
forecasted_df = pd.DataFrame(forecasted_data)

# Save the forecasted DataFrame to a CSV file
forecasted_df.to_csv('forecasted_rankings.csv', index=False)

print("Forecasted rankings saved to 'forecasted_rankings.csv'.")
```
